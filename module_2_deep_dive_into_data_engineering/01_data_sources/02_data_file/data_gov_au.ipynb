Data File - Excel

```python
import pandas as pd
import os 

# input variables
excel_file_path = "auditor_and_liquidator.xlsx"
auditor_sheetname = "REG_AUDITOR_202511"
liquidator_sheetname = "LIQUIDATOR_202510"

# output variables
# path = "data_lake" --> bị dư
```

Read Auditor data

```python
# You code here
# 1. Read data from "REG_AUDITOR_202511" spreadsheet using Pandas
df = pd.read_excel(io=excel_file_path, sheet_name=auditor_sheetname,header=2)
#df.head(5)
# 2. Apply data type conversion and export to CSV
df.dtypes # kiểm tra type của từng cột để prepare for type conversion
```

```python
# convert datatype
df = df.astype ({
    "REGISTER_NAME": "string",
    "REG_AUD_NUM": "string",
    "REG_AUD_NAME": "string",
    "REG_AUD_ACN": "string",
    "REG_AUD_STATUS": "string",
    "REG_AUD_ADD_LOCAL": "string",
    "REG_AUD_ADD_STATE": "string",
    "REG_AUD_ADD_PCODE": "string",
    "REG_AUD_ADD_COUNTRY": "string",
})

df["REG_AUD_START_DT"] = pd.to_datetime(df["REG_AUD_START_DT"], errors="coerce")
df["REG_AUD_SUSP_DT"] = pd.to_datetime(df["REG_AUD_SUSP_DT"], errors="coerce")


df.dtypes
#    "REG_AUD_ACN": "int64",
```

```python
# 3. Export to path 'destination/excel'
path = "destination/excel"
os.makedirs(path, exist_ok=True)

output_file = f"{path}/{auditor_sheetname}.csv"
print(f"Saved: {path}/{auditor_sheetname}.csv")
df.to_csv(output_file, index=False)
```

Read Liquidator data

```python
# xem dtypes ban đầu
df = pd.read_excel(io=excel_file_path, sheet_name=liquidator_sheetname,header=2)
df.dtypes
```

```python
# You code here
# 1. Read data from "LIQUIDATOR_202510" spreadsheet using Pandas
# Provide a schema when reading data file
liquidator_schema = {
    "REGISTER_NAME": str,
    "LIQ_NUM": str,
    "OFF_LIQ_NUM": str, 
    "LIQ_NAME": str,
    "LIQ_START_DT": 'datetime64[ns]',
    "OFF_LIQ_START_DT": 'datetime64[ns]',
    "LIQ_STATUS": str,
    "LIQ_SUSP_DT": 'datetime64[ns]',
    "LIQ_ADD_LOCAL": str,
    "LIQ_ADD_STATE": str,
    "LIQ_ADD_PCODE": str,
    "LIQ_ADD_COUNTRY": str,
    "LIQ_FIRM": str
}
df = pd.read_excel(io=excel_file_path, sheet_name=liquidator_sheetname, dtype = liquidator_schema)
# df.head(5)
# 2. Export to CSV as path 'destination/excel'
output_file = f"{path}/{liquidator_sheetname}.csv"
print(f"Saved: {path}/{liquidator_sheetname}.csv")
df.to_csv(output_file, index=False)
```
